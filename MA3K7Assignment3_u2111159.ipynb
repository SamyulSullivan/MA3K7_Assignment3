```python
import matplotlib.pyplot as plt
import random
def simulate_hat_numbers_for_graph(n, k):
    final_numbers_count = {}
    for _ in range(k):
        numbers = list(range(1, n + 1))
        while len(numbers) > 1:
            a, b = random.sample(numbers, 2)
            a, b = sorted((a, b))
            numbers.remove(a)
            numbers.remove(b)
            numbers.append(b - a)
        final_number = numbers[0]
        final_numbers_count[final_number] = final_numbers_count.get(final_number, 0) + 1
    return final_numbers_count

def plot_simulation_for_n_values(n_values, k):
    for n in n_values:
        result_counts = simulate_hat_numbers_for_graph(n, k)

        plt.figure(figsize=(10, 8))
        plt.bar(result_counts.keys(), result_counts.values(), color='blue')
        plt.xlabel('Final Number')
        plt.ylabel('Count')
        plt.title(f'Final Number Counts over {k} Simulations with N={n}')
        plt.xticks(range(0, n + 1))
        plt.show()

n_values_range = range(4, 15)
k_simulations = 1000
plot_simulation_for_n_values(n_values_range, k_simulations)
```
